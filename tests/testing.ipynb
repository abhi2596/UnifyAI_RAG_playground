```python
import requests 
```

```python
html_page = requests.get("https://unify.ai/benchmarks?task=text-to-image%2Ctext-generation")
```

```python
from bs4 import BeautifulSoup 

soup = BeautifulSoup(html_page.content, "html5lib")
```

```python
soup 
```

```python
model_name = []
for div in soup.find_all("div",class_="flex gap-2 items-center"):
    for element in div.children:
        model_name.append(element.get_text())
        break
    # for d in div:
    #     print(d)
```

```python
mp = {}
for name in model_name:
    html_page = requests.get(f"https://unify.ai/benchmarks/{name}")
    soup = BeautifulSoup(html_page.content, "html5lib")
    mp[name] = []
    for provider in soup.find_all("div",class_="flex-grow endpoint-list-bean endpoint-list-text inline-block"):
        mp[name].append(provider.get_text())
```

```python
mp.keys()
```

```python
import json
with open("model_provider.json","w") as f:
    f.write(json.dumps(mp))
```
